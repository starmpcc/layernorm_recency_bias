```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["axes.labelsize"] = 14
plt.rcParams["xtick.labelsize"] = 12
plt.rcParams["ytick.labelsize"] = 12
plt.rcParams["legend.fontsize"] = 12


def plot_alpha(alpha, resid):
    j = np.arange(1, 11)
    d = 64
    D_j = np.exp(np.sqrt(d)) + (j-1) * np.exp(alpha * np.sqrt(d))

    num = np.pow(D_j, 2) *(resid**2 + resid*alpha*2 + alpha) + D_j * (1-alpha) * (2*resid * np.exp(np.sqrt(d)) + np.exp(alpha* np.sqrt(d))) + (1-alpha) * (np.exp(2 * np.sqrt(d)) - np.exp(np.sqrt(d) * (1+ alpha)))
    denom = np.pow(D_j, 2)

    return j, 1/ np.sqrt(num / denom)


import matplotlib
colors = matplotlib.cm.Set1
plt.figure(figsize=(6.4, 4.2))
resid = 0
for i, alpha in enumerate(np.linspace(0, 0.8, 5)):
    plt.plot(*plot_alpha(alpha, 0), label=f"alpha={alpha:.1f}")
plt.legend()
plt.xlabel("$j$")
plt.ylabel("$h(j)$")
plt.tight_layout()
plt.savefig("../figures/alpha.pdf")
```
